1.라이브러리 및 csv파일 불러오기

```python
# numpy, pandas, matplotlib, seaborn Import 하고 csv파일 불러와서 신용점수, 연봉, 연체 정보 결측치 처리.....!
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
bank= pd.read_csv('credit_trainc.csv')
bank[["Credit Score","Annual Income","Months since last delinquent"]] = bank[["Credit Score","Annual Income","Months since last delinquent"]].fillna(bank[["Credit Score","Annual Income","Months since last delinquent"]].mean())
bank=bank.dropna()
```

```python
bank.isnull().sum()
```

```python
# 신용등급과 연봉의 상관관계
bank.plot.scatter(x='Credit Score',y='Annual Income')
```

```python
bank.head()
```

```python
#신용점수를 1000 기준으로 1과2로 할당
bank["Credit Score"][bank["Credit Score"]<1000]=1
bank["Credit Score"][bank["Credit Score"]>=1000]=2
bank.head()
```

```python
#float형 데이터 상관관계 파악
bank.corr(method="pearson").round(3)
```

```python
bank.info()
```

```python
#foalt 데이터형 산점도 전시
sns.pairplot(bank)
```

```python
#대출 상황에 따른 다른 속성들의 상관관계 산점도 
sns.pairplot(bank,y_vars="Loan Status",x_vars=["Current Loan Amount","Credit Score","Annual Income","Years in current job","Home Ownership","Purpose","Monthly Debt"])
```

```python
bank.info()
```

```python
#대출 상황에 따른 다른 속성들의 상관관계 산점도 
sns.pairplot(bank,y_vars="loan",x_vars=["age","education","housing","job","marital"])
```
